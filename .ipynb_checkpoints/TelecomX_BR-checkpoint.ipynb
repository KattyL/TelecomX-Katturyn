#游댢 Transforma칞칚o

```python
'Essa fun칞칚o extrai e transforma as chaves dos dicion치rios automaticamente em novas colunas'

def extrair_colunas(df):


    df = df.copy()
    novas_colunas = []  # guarda as colunas
    colunas_dict = []   # guarda os nomes das colunas que s칚o dicion치rios

    for coluna in df.columns:
        if df[coluna].apply(lambda x: isinstance(x, dict)).all():
            colunas_dict.append(coluna)
            colunas_expandidas = pd.json_normalize(df[coluna])
            colunas_expandidas.columns = [f"{coluna}_{nome}" for nome in colunas_expandidas.columns]
            novas_colunas.append(colunas_expandidas)

    df_sem_dicts = df.drop(columns=colunas_dict).reset_index(drop=True)

    if novas_colunas:
        df_final = pd.concat([df_sem_dicts] + novas_colunas, axis=1)
    else:
        df_final = df_sem_dicts 

    return df_final
```

### Agrupando colunas de servi칞os

Nesta etapa, estou agrupando os servi칞os contratados pelos clientes para criar uma nova classifica칞칚o (nivel_servico). Isso vai me ajudar a identificar padr칫es de comportamento em rela칞칚o  Churn e ao valor gasto, segmentando entre clientes com servi칞o completo, intermedi치rio ou b치sico. Essa simplifica칞칚o vai facilitar an치lises comparativas e destacar perfis com maior potencial de cancelamento ou reten칞칚o.

```python
colunas_servicos = [
    'phone_PhoneService',
    'phone_MultipleLines',
    'internet_InternetService',
    'internet_OnlineSecurity',
    'internet_OnlineBackup',
    'internet_DeviceProtection',
    'internet_TechSupport',
    'internet_StreamingTV',
    'internet_StreamingMovies'
]

def contar_servicos_ativos(linha):
    return sum(
        1 for col in colunas_servicos
        if linha[col] not in ['No', 'Sem servi칞o', 'No internet service']
    )

df['qtd_servicos_ativos'] = df.apply(contar_servicos_ativos, axis=1)

def classificar_cliente(qtd):
    if qtd >= 6:
        return 'Servi칞o completo'
    elif qtd >= 3:
        return 'Servi칞os intermedi치rios'
    else:
        return 'Servi칞os b치sicos'

df['nivel_servico'] = df['qtd_servicos_ativos'].apply(classificar_cliente)
```
