#üîß Transforma√ß√£o

```python
'Essa fun√ß√£o extrai e transforma as chaves dos dicion√°rios automaticamente em novas colunas'

def extrair_colunas(df):


    df = df.copy()
    novas_colunas = []  # guarda as colunas
    colunas_dict = []   # guarda os nomes das colunas que s√£o dicion√°rios

    for coluna in df.columns:
        if df[coluna].apply(lambda x: isinstance(x, dict)).all():
            colunas_dict.append(coluna)
            colunas_expandidas = pd.json_normalize(df[coluna])
            colunas_expandidas.columns = [f"{coluna}_{nome}" for nome in colunas_expandidas.columns]
            novas_colunas.append(colunas_expandidas)

    df_sem_dicts = df.drop(columns=colunas_dict).reset_index(drop=True)

    if novas_colunas:
        df_final = pd.concat([df_sem_dicts] + novas_colunas, axis=1)
    else:
        df_final = df_sem_dicts 

    return df_final
```

## Agrupando colunas de servi√ßos

Nesta etapa, estou agrupando os servi√ßos contratados pelos clientes para criar uma nova classifica√ß√£o (nivel_servico). Isso vai me ajudar a identificar padr√µes de comportamento em rela√ß√£o √† Churn e ao valor gasto, segmentando entre clientes com servi√ßo completo, intermedi√°rio ou b√°sico. Essa simplifica√ß√£o vai facilitar an√°lises comparativas e destacar perfis com maior potencial de cancelamento ou reten√ß√£o.

```python
colunas_servicos = [
    'phone_PhoneService',
    'phone_MultipleLines',
    'internet_InternetService',
    'internet_OnlineSecurity',
    'internet_OnlineBackup',
    'internet_DeviceProtection',
    'internet_TechSupport',
    'internet_StreamingTV',
    'internet_StreamingMovies'
]

def contar_servicos_ativos(linha):
    return sum(
        1 for col in colunas_servicos
        if linha[col] not in ['No', 'Sem servi√ßo', 'No internet service']
    )

df['qtd_servicos_ativos'] = df.apply(contar_servicos_ativos, axis=1)

def classificar_cliente(qtd):
    if qtd >= 6:
        return 'Servi√ßo completo'
    elif qtd >= 3:
        return 'Servi√ßos intermedi√°rios'
    else:
        return 'Servi√ßos b√°sicos'

df['nivel_servico'] = df['qtd_servicos_ativos'].apply(classificar_cliente)
```

```python
df.groupby(['nivel_servico', 'Churn']).size().unstack()
```

# üìä Carga e an√°lise

```python
df_churn = df.groupby('nivel_servico').agg({
    'Churn': ['sum', 'count']
})

df_churn.columns = ['churn_count', 'total_count']
df_churn['razao_churn_percent'] = (df_churn['churn_count'] / df_churn['total_count']) * 100
df_churn_sorted = df_churn.sort_values('razao_churn_percent')

plt.figure(figsize=(8, 5))
sns.barplot(x=df_churn_sorted.index, y=df_churn_sorted['razao_churn_percent'], palette='viridis_r')
plt.ylabel('Taxa de Churn (%)')
plt.xlabel('N√≠vel de Servi√ßo')
plt.title('Taxa de Churn por N√≠vel de Servi√ßo')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### An√°lise de Churn e Pre√ßo por N√≠vel de Servi√ßo

Observamos que o √≠ndice de churn para clientes dos **servi√ßos intermedi√°rios** √© ***alarmantemente alto (62%)***, muito superior aos **servi√ßos completos (25%)** e **b√°sicos (22%)**. Esse dado sugere uma forte insatisfa√ß√£o ou desconex√£o do cliente com o pacote intermedi√°rio.

Analisando o pre√ßo m√©dio mensal, temos:

| **N√≠vel de Servi√ßo**         | **Pre√ßo M√©dio (R$)** |

Curiosamente, o pacote intermedi√°rio apresenta um pre√ßo m√©dio consideravelmente elevado, muito pr√≥ximo ao pacote completo, mas com um churn que supera em mais do que o dobro o dos servi√ßos completos.

**Recomendo uma revis√£o urgente no posicionamento, oferta e comunica√ß√£o do pacote intermedi√°rio**, buscando aumentar seu valor percebido, melhorar a experi√™ncia do cliente, ou ajustar a precifica√ß√£o para reduzir esse churn cr√≠tico.


```python
churn_contrato = pd.crosstab(
    df['account_Contract'], df['Churn'], normalize='index') * 100
churn_contrato.columns = ['N√£o', 'Sim']

color_nao = '#5bc561'
color_sim = '#ed4039' 

ax = churn_contrato.plot(
    kind='bar',
    stacked=True,
    color=[color_nao, color_sim],
    figsize=(6, 4)
)

traducao_contratos = {
    'Month-to-month': 'Mensal',
    'One year': 'Anual',
    'Two year': 'Bienal'
}

r√≥tulos_traduzidos = [traducao_contratos.get(label.get_text(), label.get_text()) for label in ax.get_xticklabels()]
ax.set_xticklabels(r√≥tulos_traduzidos, rotation=0)

ax.set_title('Churn (%) por Tipo de Contrato', fontsize=14, fontweight='bold')
ax.set_ylabel('Percentual (%)', fontsize=12)
ax.set_xlabel('Tipo de Contrato', fontsize=12)
ax.legend(title='Churn', loc='upper left', bbox_to_anchor=(1,1))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
